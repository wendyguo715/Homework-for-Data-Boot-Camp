```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player = len(purchase_data["SN"].value_counts())
player_total = pd.DataFrame([player], columns = ["Total Players"])
player_total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of Unique Items
item_ID = len(purchase_data["Item ID"].value_counts())

# Average Purchase Price
aveprice = purchase_data["Price"].mean()

# Total Number of Purchases
num_purchases = len(purchase_data["Purchase ID"])

# Total Revenue
sum_rev = purchase_data["Price"].sum()

Pur_Analysis_Total = pd.DataFrame({"Number of Unique Items": [item_ID],
                                    "Average Purchase Price": ['${:,.2f}'.format(aveprice)],
                                    "Number of Purchases": [num_purchases],
                                    "Total Revenue": ['${:,.2f}'.format(sum_rev)]})

Pur_Analysis_Total
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Grouping by gender & counting
df_gender = purchase_data[["SN", "Gender"]]
gender = df_gender.drop_duplicates()
counts = gender["Gender"].value_counts()

total_counts = [counts[0],counts[1],counts[2]]
percents = [round((counts[0]/player)*100,2),round((counts[1]/player)*100,2),round((counts[2]/player)*100,2)]

gender_summary = pd.DataFrame({"Total Count": total_counts, "Percentage of Players": percents})
gender_summary.index = (["Male", "Female", "Other / Non-Disclosed"])
gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pur_analysis_gender = purchase_data.groupby('Gender')
purchase_count = pur_analysis_gender["Purchase ID"].count()
ave_purchase = pur_analysis_gender["Price"].mean()
sum_purchase = pur_analysis_gender["Price"].sum()
ave_purchase_person = sum_purchase/counts

pur_analysis_gender_df = pd.DataFrame({"Purchase Count": purchase_count,
                                       "Average Purchase Price": ave_purchase,
                                       "Total Purchase Value": sum_purchase,
                                       "Avg Total Purchase per Person": ave_purchase_person
                                      })

pur_analysis_gender_df["Average Purchase Price"] = pur_analysis_gender_df["Average Purchase Price"].map("${:.2f}".format)
pur_analysis_gender_df["Total Purchase Value"] = pur_analysis_gender_df["Total Purchase Value"].map("${:.2f}".format)
pur_analysis_gender_df["Avg Total Purchase per Person"] = pur_analysis_gender_df["Avg Total Purchase per Person"].map("${:.2f}".format)
                            
pur_analysis_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

df_age = purchase_data[["SN", "Age"]]
age = df_age.drop_duplicates()
age_bins_groups = pd.cut(age["Age"], age_bins, labels=group_names)

age_group = age.groupby(age_bins_groups)
age_group_count = age_group["Age"].count()
age_group_percent = round(age_group_count/player*100,2)

age_group_count_df = pd.DataFrame({"Total Count":age_group_count,"Percentage of Players": age_group_percent})
age_group_count_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins_groups_2 = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
age_group = purchase_data.groupby(age_bins_groups_2)

age_purchase_count = age_group["Age"].count()
age_ave_purchase = age_group["Price"].mean()
age_sum_purchase = age_group["Price"].sum()
age_ave_purchase_person = age_sum_purchase/age_group_count

pur_analysis_age_df = pd.DataFrame({"Purchase Count": age_purchase_count,
                                       "Average Purchase Price": age_ave_purchase,
                                       "Total Purchase Value": age_sum_purchase,
                                       "Avg Total Purchase per Person": age_ave_purchase_person
                                      })

pur_analysis_age_df["Average Purchase Price"] = pur_analysis_age_df["Average Purchase Price"].map("${:.2f}".format)
pur_analysis_age_df["Total Purchase Value"] = pur_analysis_age_df["Total Purchase Value"].map("${:.2f}".format)
pur_analysis_age_df["Avg Total Purchase per Person"] = pur_analysis_age_df["Avg Total Purchase per Person"].map("${:.2f}".format)

pur_analysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender = purchase_data.groupby("SN")

toper_purchase_count = top_spender["Purchase ID"].count()
toper_ave_purchase = top_spender["Price"].mean()
toper_sum_purchase = top_spender["Price"].sum()

toper_df = pd.DataFrame({"Purchase Count": toper_purchase_count,
                         "Average Purchase Price": toper_ave_purchase,
                         "Total Purchase Value": toper_sum_purchase,
                         })
toper_df_sort = toper_df.nlargest(5, "Total Purchase Value")

toper_df_sort["Average Purchase Price"] = toper_df_sort["Average Purchase Price"].map("${:.2f}".format)
toper_df_sort["Total Purchase Value"] = toper_df_sort["Total Purchase Value"].map("${:.2f}".format)

toper_df_sort
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_items = purchase_data.groupby(["Item ID","Item Name"])

pop_item_count = pop_items["Purchase ID"].count()
pop_item_price = pop_items["Price"].mean()
pop_item_sum = pop_items["Price"].sum()

pop_items_df = pd.DataFrame({"Purchase Count": pop_item_count,
                             "Item Price": pop_item_price,
                             "Total Purchase Value": pop_item_sum,
                             })

pop_items_df_sort = pop_items_df.sort_values("Purchase Count", ascending=False).head()

pop_items_df_sort["Item Price"] = pop_items_df_sort["Item Price"].map("${:.2f}".format)
pop_items_df_sort["Total Purchase Value"] = pop_items_df_sort["Total Purchase Value"].map("${:.2f}".format)

pop_items_df_sort
```
